## Plot a histogram

### All species

````python
penguins = sns.load_dataset("penguins")
sns.displot(penguins, x="flipper_length_mm")
````

### Colour by species

````python
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")
````

### Normalise data

````python
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="density", common_norm=False, element="step")
````

### Smoothed data

````python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill=True)
````

````python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
````

````python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
````

````python
sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde"
)
````

````python
g = sns.PairGrid(penguins)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
````

````python
titanic = sns.load_dataset("titanic")
titanic
````

````python
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic, capsize = 0.05, errwidth = 2)
````

````python
iris = sns.load_dataset("iris")
sns.catplot(data=iris, orient="h", kind="box", height=5, aspect = 2)
````

````python
sns.violinplot(x=iris.species, y=iris.sepal_length)
````

````python
g = sns.catplot(x="fare", y="survived", row="class",
                kind="box", orient="h", height=3, aspect=4,
                data=titanic.query("fare > 0"))
g.set(xscale="log")
````

````python
tips = sns.load_dataset("tips")
tips
````

````python
sns.regplot(x="total_bill", y="tip", data=tips)
````

## Heatmap

````python
sns.heatmap(penguins.corr(), annot=True, cmap='RdYlGn')
plt.title("Correlation matrix");
````

````python
import ppscore as pps
pps_matrix = pps.matrix(penguins)
pps_pivot = pps_matrix.pivot('x', 'y', 'ppscore')
pps_pivot.index.name, pps_pivot.columns.name = None, None

sns.heatmap(pps_pivot, annot=True, cmap='YlGn', annot_kws={"fontsize":10})
plt.title("Predictive Power Score Matrix");
````

````python
import plotly.express as px
df = px.data.iris()
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_width',
              color='species', size='petal_length', size_max=22, opacity=0.7, width=950, height=950)
fig.update_layout(
    margin=dict(l=0, r=0, t=0, b=0),)
````
